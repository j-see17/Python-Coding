4. What is the **loading** between variable **`Q4: D`** and the **second principal component**? (Hint: You are asked about the loading, not about the coefficient of the principal direction!)


```python
# Hint:
pca_directions = pd.DataFrame(
    pca.components_.T,
    index=X.columns,
    columns=['PC' + str(l + 1) for l in range(pca.n_components_)],
)


from sklearn.decomposition import PCA
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/brunojacobs/BUMK744/master/data/car_survey.csv'
X = pd.read_csv(url)

# Standardize exactly like the notebook
X = (X - X.mean()) / X.std()

# Fit PCA
pca = PCA()
pca.fit(X)

# Principal directions
pca_directions = pd.DataFrame(
    pca.components_.T,
    index=X.columns,
    columns=['PC' + str(i + 1) for i in range(pca.n_components_)]
)

# Loadings = direction coefficients * sqrt(eigenvalue)
loadings = pca_directions.mul(np.sqrt(pca.explained_variance_), axis=1)

answer = loadings.loc['Q4: D', 'PC2']
print(round(answer, 4))
```
