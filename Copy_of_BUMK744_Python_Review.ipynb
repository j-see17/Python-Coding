## **[B] The print function**

The canonical way to get introduced to any programming language is by writting a program that outputs "Hello, World!" to the screen.

> - In programming, **outputting** something to the screen is known as **printing**
> - In Python, you **print** something to the screen by *calling* (using) the **print function**

````python
print('Hello, World!')
````

Let's dissect this line of code, piece by piece:
- **`print`** is the **name** of the function we are calling
- **`()`** are **parentheses** (sometimes called round brackets) <br>
  - The parentheses following a function name can be filled with **arguments**
- **`'Hello, World!'`** is the single **argument** we provide to the print function <br>
  - Think of arguments as *inputs* to a function

Note that the argument **`'Hello, World!'`** is surrounded by single quotes **`' '`**
> - In this case, the argument **`'Hello, World!'`** is called a **string** <br>
A **string** is a sequence of characters, i.e. a text snippet
> - Python needs the single quotes **`' '`** to distinguish text from code
> - In Python, you can use either single quotes **`' '`** or double quotes **`" "`** to indicate a string. There is no real difference, it is a matter of preference
> - Note that **the quotes are not printed** as they are not part of the string, they are only used to indicate to Python that we are providing a string

Note: Python is very versatile in handling strings (that is, text data), it is one of its strengths!

## **[F] Conditional statements**

A basic computer program executes each line, after line, after line, ..., in a predefined order.

> **Control flow** allows you to change the order in which your computer program is executed
<br>
> The order is now made conditional on values of your variables

Today we will first look at **conditional statements**, but later on during the course we will explore **`for`** loops and **`while`** loops as well.

### **`if`**

A block of code under **`if STATEMENT:`** syntax is only executed if the **`STATEMENT`** is **`True`**

> * Note the **indentation** (whitespace) of the block under the **`if`** statements!
> * You create an indentation by hitting the **tab** button **inside** the code cell
> * This use of whitespace is one of the more unique features of Python
> * Compare this against R, where you use curly braces `{ }` to indicate a block of code

````python
x = 3
y = 4

if x == y:
  print('x and y the same')
  print('x=y:', x)

if x != y:
  print('x and y are not the same')
  print('x:', x)
  print('y:', y)
````

As a special case we can use **numerical variables** directly in a conditional statement.

> Let **`x`** be a number
> * If **`x`** is **nonzero** it is considered **True** in a conditional statement
> * If **`x`** is **zero** it is considered **False** in a conditional statement
> * In essence, **`if x:`** is equavalent to **`if x != 0:`**

````python
x = 5
if x:
  print('This prints if x is not equal to zero:', x)

y = 0
if y:
  print('This does not print if y is equal to zero:', y)

z = 3.14
if z:
  print(z)
````

### **`if`**, **`else`**

The **`if STATEMENT:`** syntax can be extended using an **`else:`** statement.

The block of code under the **`else:`** will only be executed if **`STATEMENT`** is **`False`**:

````python
course_number = 123

if course_number == 744:
  print('My favorite course is:', course_number, '(but you knew that already)')
else:
  print(course_number, 'is definitely a fun course, but not my favorite')
````

### **`if`**, **`elif`**, **`else`**

The **`if STATEMENT:`** syntax can be further extended by using an **`elif OTHER_STATEMENT:`** statement.

The block of code under the **`elif OTHER_STATEMENT:`** will only be executed if **`STATEMENT`** is **`False`** and **`OTHER_STATEMENT:`** is **`True`**:

The block of code under the **`else:`** will only be executed if both **`STATEMENT`** and **`OTHER_STATEMENT:`** are **`False`**:

> About **`elif`**:
> - It is always placed *after* **`if`**, but *before* an **`else`**
> - You can have *more than one* **`elif`**

````python
course_number = 123

if course_number == 744:
  print('I teach', course_number, 'in the MSMA program')
elif course_number == 452:
  print('I teach', course_number, 'in the UG program')
elif course_number == 858:
  print('I teach', course_number, 'in the PhD program')
else:
  print('I do not teach', course_number)
````

## **[G] Packages and modules**

**Packages** (also known as modules) *significantly* extend the functionality of Python

> *A Python package (module) is similar to an R package*

Two types of packages in Python:
* **Internal**: Always available (also called built-in modules)
* **External**: Separate packages that (sometimes) need to be installed

Examples of popular (external) data science packages:
> * `numpy`
> * `pandas`
> * `scipy`
> * `scikit-learn`

In this course you will become familiar with the packages mentioned above.
> * Although these packages are external, they come pre-installed on Google Colab. <br> **You do not have to install anything yourself!**

### Loading a Package

A package can be loaded using the **`import`** statement. After a package has been *imported* you can access its functionality.

There are **4 options to importing a package**:
> 1. Explicit import of a package
> 2. Explicit import of a package by alias
> 3. Explicit import of **some** of a package's contents
> 4. *Implicit* import of **all** of a package's contents

Only options 1 and 2 should be used. Option 3 is fine, but is somewhat of limited use in practice. **Option 4 should not be used**.

### 1. Explicit import of package
Using the **`import PACKAGE`** syntax:


````python
import math
````

After the `math` module has been explicitly imported, you can access its contents using `math.CONTENT`.

Example 1:
> The square root function `sqrt()` is now available through the `math` module

````python
math.sqrt(9)
````

Example 2:
> The value of pi (3.1415...) `pi` can be accessed through the `math` module

````python
math.pi
````

### 2. Explicit import of a package by alias
Using the **`import PACKAGE as ALIAS`** syntax:

````python
import math as m
````

The contents of the `math` module can now be accessed using `m.CONTENT`.

> The `sqrt()` function in `math` can now be accessed as `m.sqrt()`

````python
m.sqrt(16)
````

### 3. Explicit import of **some** of a package's contents
Using the **`from PACKAGE import CONTENT_A, CONTENT_B`** syntax:

````python
from math import log, pi, e
````

In the cell above we have imported the following contents from the `math` module:
* Function `log()` (natural logarithm)
* Values `pi` and `e` (Euler's number)

This content can now be accessed directly, without having to prepend with `math.` or `m.`


````python
print(pi)
print(log(744))
print(log(e))
````

### 4. Implicit import of **all** of a package's contents (**not recommended!**)
Using the **`from PACKAGE import *`** syntax:

````python
from math import *
````

Here the asterisk (`*`) means that all the content of the `math` module can now be accessed directly, without prepending with `math.`

````python
print(cos(1.0))
print(sin(1.0))
print(sqrt(744))
````

### dict
A dictionary is a **mapping** of **keys** to **values**

**keys**
- Think of a **key** as an index, such as we have seen for a list and a tuple
- The difference is that the keys of a dictionary are specified by you: they are not determined by the order of entry.

**values**
- A **value** in a dictionary can be accessed by using its corresponding **key**

###Pseudo random

The random numbers that we generate are actually known as **pseudo random numbers** because they are not truly random. We can control the **state** of the random number generator such that our results become **reproducible**. **This is a very important property in data science, as you should be able to replicate your results**.

The way to control the **state** of the random number generator in NumPy is by specifing a **seed number** to the **`np.random.default_rng`** function:

````python
# WITHOUT specifying a seed:
rng_a = np.random.default_rng()
print(rng_a.uniform())

rng_b = np.random.default_rng()
print(rng_b.uniform())

# random numbers from rng_a and rng_b are not the same
````

````python
# WITH specifying a seed:

rng_a = np.random.default_rng(seed=744)
print(rng_a.uniform())

rng_b = np.random.default_rng(seed=744)
print(rng_b.uniform())

# random numbers from rng_a and rng_b now are the same
````
