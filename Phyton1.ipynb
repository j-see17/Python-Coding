```python
# data analysis
import pandas as pd
data = pd.read_csv(r"data_wendys.csv")


# data cleaning
data_clean = data.dropna(subset=['patronage_wendys','rating_wendys'])
rating_wendys = data_clean['rating_wendys']
rating_wendys.plot.hist()

# chi-square test

# H0: there is no relationship between education and income
# HA: there is a relationship between education and income
tab = pd.crosstab(data_clean['income'],data_clean['edu'])

# additional tools to import for chi-squared tests
from scipy.stats import chi2_contingency
c, p, dof, expected = chi2_contingency(tab)
print
```
